## Data cleaning

### Subtask:
Handle potential outliers in the numerical features of the dataset.


**Reasoning**:
Identify outliers in the numerical features using box plots and handle them by winsorizing the extreme values at the 99th percentile. Then verify the effectiveness of the outlier handling.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a copy of the DataFrame to avoid modifying the original
df_cleaned = df.copy()

# Identify numerical features (excluding 'Class')
numerical_features = df_cleaned.select_dtypes(include=['number']).columns

# Winsorize outliers at the 99th percentile for selected features
for col in ['AREA', 'PERIMETER', 'MAJOR_AXIS']:
    upper_limit = df_cleaned[col].quantile(0.99)
    df_cleaned[col] = df_cleaned[col].clip(upper=upper_limit)

# Verify the effectiveness of the outlier handling
plt.figure(figsize=(15, 5))
for i, feature in enumerate(['AREA', 'PERIMETER', 'MAJOR_AXIS']):
    plt.subplot(1, 3, i + 1)
    sns.boxplot(y=df_cleaned[feature], color=f'C{i}')
    plt.title(f'Boxplot of {feature} after Winsorization')
plt.tight_layout()
plt.show()

print(f"Shape of the DataFrame after outlier handling: {df_cleaned.shape}")
```

## Feature engineering

### Subtask:
Engineer new features from existing ones in the cleaned dataset, `df_cleaned`, focusing on ratios and combinations that might improve model performance.


**Reasoning**:
Create new features based on ratios and combinations of existing features in `df_cleaned`, then evaluate these features and store the updated dataframe.



```python
# Create ratio features
df_cleaned['Width_to_Length'] = df_cleaned['MINOR_AXIS'] / df_cleaned['MAJOR_AXIS']
df_cleaned['Area_to_Perimeter'] = df_cleaned['AREA'] / df_cleaned['PERIMETER']
df_cleaned['Convex_to_Area'] = df_cleaned['CONVEX_AREA'] / df_cleaned['AREA']

# Create interaction features
df_cleaned['Area_Perimeter_Interaction'] = df_cleaned['AREA'] * df_cleaned['PERIMETER']

# Evaluate new features
import matplotlib.pyplot as plt
import seaborn as sns

# Distributions
plt.figure(figsize=(15, 5))
for i, feature in enumerate(['Width_to_Length', 'Area_to_Perimeter', 'Convex_to_Area', 'Area_Perimeter_Interaction']):
  plt.subplot(1, 4, i + 1)
  sns.histplot(df_cleaned[feature], kde=True)
  plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()

# Correlations of numerical features
numerical_df = df_cleaned.select_dtypes(include=['number'])
correlations_numerical = numerical_df.corr()
print(correlations_numerical)

# Correlation of new features with Class
from scipy.stats import pointbiserialr

new_features = ['Width_to_Length', 'Area_to_Perimeter', 'Convex_to_Area', 'Area_Perimeter_Interaction']
for feature in new_features:
  correlation, p_value = pointbiserialr(df_cleaned[feature], df_cleaned['Class'].astype('category').cat.codes)
  print(f"Point-biserial correlation between {feature} and Class: {correlation}, p-value: {p_value}")

display(df_cleaned.head())
```

## Model training

### Subtask:
Train a RandomForestClassifier on the training data (`X_train`, `y_train`)


**Reasoning**:
Train a RandomForestClassifier on the training data (X_train, y_train).



```python
from sklearn.ensemble import RandomForestClassifier

# Instantiate a RandomForestClassifier
rf_model = RandomForestClassifier(random_state=42)

# Train the classifier
rf_model.fit(X_train, y_train)
```

## Model evaluation

### Subtask:
Evaluate the trained RandomForestClassifier model (`rf_model`) using the test data (`X_test`, `y_test`).


**Reasoning**:
Evaluate the trained RandomForestClassifier model using the test data and visualize the results.



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Predict using the trained model
y_pred = rf_model.predict(X_test)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
cm = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")
print("\nConfusion Matrix:")
print(cm)

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Visualize the confusion matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=rf_model.classes_, yticklabels=rf_model.classes_)
plt.title("Confusion Matrix Heatmap")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```
